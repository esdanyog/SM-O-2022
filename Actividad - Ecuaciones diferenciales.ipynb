# Actividad - Ecuaciones diferenciales

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from scipy.integrate import solve_ivp, odeint
from ipywidgets import *
```

1. Resolver numéricamente con `solve_ivp` la siguiente ecuación diferencial:
$$\frac{dx}{dt}=\frac{x}{t+1}$$

```python
def f1(t,x):
    return x/(t+1)

tspan = [0, 50]
x0 = np.array([2])
x1 = solve_ivp(f1, tspan, x0, max_step=0.1)
```

Grafique con condición inicial $x_0 = 2$

```python
plt.figure()
plt.plot(x1.t, x1.y[0, :], "r", label="Solución Numérica")
plt.legend(loc="best")
```
